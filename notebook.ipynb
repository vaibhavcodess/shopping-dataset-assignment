```python
import pandas as pd

df = pd.read_csv("Combined_dataset.csv")

df.head()
```

```python
# Shape of dataset
print("Shape of dataset:", df.shape)

# Column names
print("\nColumns:")
print(df.columns)

# Data types
print("\nData Types:")
print(df.dtypes)

# First 5 rows
print("\nFirst 5 Rows:")
print(df.head())

# Last 5 rows
print("\nLast 5 Rows:")
print(df.tail())
```

```python
# Checking missing values
print(df.isnull().sum())
```

```python
# Fill missing values

df['discount'] = df['discount'].fillna(0)

df['what_customers_said'] = df['what_customers_said'].fillna('Not Available')

df['seller_name'] = df['seller_name'].fillna('Not Available')

df['videos'] = df['videos'].fillna('Not Available')

df['seller_information'] = df['seller_information'].fillna('Not Available')

df['variations'] = df['variations'].fillna('Not Available')

print("Missing values handled successfully!")
```

```python
# Checking missing values again

print(df.isnull().sum())
```

```python
# Check duplicate rows

print("Duplicate rows before removing:", df.duplicated().sum())

# Remove duplicates
df = df.drop_duplicates()

# Check again
print("Duplicate rows after removing:", df.duplicated().sum())
```

```python
# Cleaning final_price column

df['final_price'] = df['final_price'].replace('[₹,"]', '', regex=True)

df['final_price'] = df['final_price'].astype(float)

print(df['final_price'].head())
```

```python
# Creating quantity column

df['quantity'] = 1

# Creating total_amount column

df['total_amount'] = df['final_price'] * df['quantity']

# Display first 5 rows

print(df[['title', 'final_price', 'quantity', 'total_amount']].head())
```

```python
# Save cleaned dataset

df.to_csv("cleaned_dataset.csv", index=False)

print("Cleaned dataset saved successfully!")
```
